```python
def process_and_overwrite_faces(image_path, size=(224, 224)):
    img = cv2.imread(image_path)
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    
    faces = face_cascade.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
    if len(faces) > 0:
        for (x, y, w, h) in faces:
          
            face = gray_img[y:y+h, x:x+w]  

            
            resized_face = cv2.resize(face, size)

          
            cv2.imwrite(image_path, resized_face)
            print(f"Overwritten image: {image_path}")
            break
    else:
        os.remove(image_path)

# Function to iterate through folders and overwrite the images
def process_faces_in_folders(root_folder, size=(224, 224)):
    for foldername, subfolders, filenames in os.walk(root_folder):
        for filename in filenames:
            if filename.endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tif')):
                file_path = os.path.join(foldername, filename)

                # Process and overwrite the image with the detected face
                process_and_overwrite_faces(file_path, size)
```

```python
input_folder = "C:/Users/Ahmed.Kirat/Desktop/test"  # Replace with your folder path
output_folder = "C:/Users/Ahmed.Kirat/Desktop/test" # Replace with the path where you want to save the processed images
image_size = (224, 224)  # Set the size for the DNN input, e.g., 224x224

process_faces_in_folders(input_folder, image_size)
print(f"Processing and overwriting images in '{input_folder}'.")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
from keras.callbacks import LearningRateScheduler


train_dir = "C:/Users/Ahmed.Kirat/Desktop/test"


train_datagen = ImageDataGenerator(rescale=1./255, horizontal_flip=True, zoom_range=0.2)


batch_size = 32
image_size = (150, 150)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='binary'  
)

def lr_scheduler(epoch, lr):
        return lr * np.exp(-0.1)

model = Sequential()


model.add(Flatten(input_shape=(150, 150, 3)))



model.add(Dense(256, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(64, activation='relu'))






model.add(Dense(1, activation='sigmoid'))  

lr_schedule = LearningRateScheduler(lr_scheduler)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])


```

```python
history = model.fit(
    train_generator,
    epochs=50, callbacks=[lr_schedule] 
)
```

```python
print("Accuraccy:", model.evaluate(train_generator))
```
